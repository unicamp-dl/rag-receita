```python
from google.colab import drive, files
import os
import tiktoken

drive.mount('/content/drive', force_remount=True)
```

```python
# Caminho da pasta raiz
base_path = "/content/drive/Shareddrives/RAG Receita/Scraping Acórdãos/Acórdãos txt/"

# Inicializar tokenizador
tokenizer = tiktoken.get_encoding("cl100k_base")  # Encoding usado no text-embedding-3-small

# Preço do modelo (USD)
COST_PER_1K_TOKENS = 0.00002

# Contagem total de tokens
total_tokens = 0

# Percorrer as pastas dentro de "Acórdãos txt"
for folder in os.listdir(base_path):
    folder_path = os.path.join(base_path, folder)

    # Garantir que é uma pasta e começa com número + string
    if os.path.isdir(folder_path) and folder[0].isdigit():
        year_path = os.path.join(folder_path, "2023")  # Pasta específica de 2023

        if os.path.isdir(year_path):
            for file_name in os.listdir(year_path):
                if file_name.endswith(".txt"):
                    file_path = os.path.join(year_path, file_name)

                    # Ler o conteúdo do arquivo
                    with open(file_path, "r", encoding="utf-8") as f:
                        text = f.read()

                    # Contar tokens
                    num_tokens = len(tokenizer.encode(text))
                    total_tokens += num_tokens

# Calcular custo total
total_cost = (total_tokens / 1000) * COST_PER_1K_TOKENS

# Exibir resultados
print(f"Total de tokens: {total_tokens}")
print(f"Custo estimado: ${total_cost:.6f}")
```
